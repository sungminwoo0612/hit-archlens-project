## ë°©ë²• 2: Pythonì„ ì‚¬ìš©í•œ ë‹¤ìš´ë¡œë“œ


```python
import requests
from tqdm import tqdm

def download_file(url: str, output_path: Path, chunk_size: int = 8192):
    """íŒŒì¼ ë‹¤ìš´ë¡œë“œ (ì§„í–‰ë¥  í‘œì‹œ)"""
    response = requests.get(url, stream=True)
    total_size = int(response.headers.get('content-length', 0))
    
    with open(output_path, 'wb') as f, tqdm(
        desc=output_path.name,
        total=total_size,
        unit='B',
        unit_scale=True,
        unit_divisor=1024,
    ) as pbar:
        for chunk in response.iter_content(chunk_size=chunk_size):
            if chunk:
                f.write(chunk)
                pbar.update(len(chunk))
    
    print(f"âœ… ë‹¤ìš´ë¡œë“œ ì™„ë£Œ: {output_path}")

# ì‚¬ìš© ì˜ˆì‹œ (ì£¼ì„ ì²˜ë¦¬)
# AWS_URL = "https://d1.awsstatic.com/webteam/architecture-icons/q1-2024/Asset-Package_01242024.7c4f8b8b.zip"
# ZIP_PATH = DATA_DIR / "Asset-Package.zip"
# download_file(AWS_URL, ZIP_PATH)
```

```python
def extract_zip(zip_path: Path, extract_to: Path):
    """ZIP íŒŒì¼ ì••ì¶• í•´ì œ"""
    if not zip_path.exists():
        print(f"âŒ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {zip_path}")
        return
    
    print(f"ğŸ“¦ ì••ì¶• í•´ì œ ì¤‘: {zip_path.name}")
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)
    print(f"âœ… ì••ì¶• í•´ì œ ì™„ë£Œ: {extract_to}")

# ZIP íŒŒì¼ ì°¾ê¸°
zip_files = list(DATA_DIR.glob("Asset-Package*.zip"))
if zip_files:
    print(f"ë°œê²¬ëœ ZIP íŒŒì¼: {len(zip_files)}ê°œ")
    for zip_file in zip_files:
        print(f"  - {zip_file.name}")
        # ì´ë¯¸ ì••ì¶• í•´ì œëœ ë””ë ‰í„°ë¦¬ê°€ ìˆëŠ”ì§€ í™•ì¸
        extract_dir = DATA_DIR / zip_file.stem
        if not extract_dir.exists():
            extract_zip(zip_file, DATA_DIR)
        else:
            print(f"  ì´ë¯¸ ì••ì¶• í•´ì œë¨: {extract_dir}")
else:
    print("âš ï¸  ZIP íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ë¨¼ì € ë‹¤ìš´ë¡œë“œí•˜ì„¸ìš”.")
```
